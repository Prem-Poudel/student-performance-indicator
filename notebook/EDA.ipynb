# Student Performance Indicator

Data Source: https://www.kaggle.com/datasets/spscientist/students-performance-in-exams?datasetId=74977
This is a student performance indicator. It is used to measure student performance. It is a simple Machine Learning model. The dataset is from Kaggle.
The dataset has 1000 rows and 8 columns.
The Dataset contains the following columns:

```python
['Gender', 'Race/Ethnicity', 'Parental Level of Education', 'Lunch', 'Test Preparation Course', 'Math score', 'Reading score', 'Writing score']
```


## 1. Imports

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
print("setup completed")
````

### 1.1. Dataset

````python
file_path = "data/StudentsPerformance.csv"
df = pd.read_csv(file_path)
````

````python
df.head()
````

### About the dataset

````python
def line_space():
    print("------"*8)

print("Dataset size:\t", df.shape)
line_space()
print("Columns:\t", df.columns)
line_space()
print("Index:\t", df.index)
line_space()
print("Data types:\t", df.dtypes)
line_space()
print("Null values:\t", df.isnull().sum())
line_space()
print("Duplicate values:\t", df.duplicated().sum())
line_space()
print("Unique values:\t", df.nunique())
````

### 1.2. Dataset description

````python
df.describe()
````

### Adding Total Score and Average Score column

````python
df["total_score"] = df["math score"] + df["reading score"] + df["writing score"]
df["average_score"] = df["total_score"] / 3
````

````python
df.head()
````

## 2. Visualization
- Histograms
- KDE

````python
fig, ax = plt.subplots(1,2,figsize=(10, 6))

sns.histplot(data=df,x='average_score',bins=30,kde=True,color='g', ax=ax[0])

sns.histplot(data=df,x='average_score',kde=True,hue='gender', ax=ax[1])
plt.show()
````

````python
fig, ax = plt.subplots(1,3,figsize=(15, 6))
sns.histplot(data=df,x='average_score',kde=True,hue='lunch', ax=ax[0])
sns.histplot(data=df[df.gender=='female'],x='average_score',kde=True,hue='lunch', ax=ax[1])
sns.histplot(data=df[df.gender=='male'],x='average_score',kde=True,hue='lunch', ax=ax[2])
plt.show()
````
